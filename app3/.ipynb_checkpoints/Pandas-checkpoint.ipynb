```python
## Geocoding
```

```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
```

```python
n=nom.geocode("735 Dolores St, San Francisco, CA 94119")
```

```python
import pandas as pd
df7=pd.read_csv("./csv/supermarkets.csv")

df7["Address"]=df7["Address"]+","+ df7["City"] + ","+ df7["State"]+ "," + df7["Country"]
df7
```

```python
df7["Coordinates"]=df7["Address"].apply(nom.geocode)
df7.Coordinates[0].latitude
```

```python
df7.Coordinates[0].longitude
```

```python
df7["Latitude"]=df7["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df7["Longitude"]=df7["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df7
```
